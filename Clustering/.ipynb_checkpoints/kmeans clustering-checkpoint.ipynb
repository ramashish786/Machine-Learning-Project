```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.cluster import KMeans
```

```python
unclean_data = pd.read_csv('D:/machine_learning/Data/airbnb.csv') 
```

```python
unclean_data.head()
```

```python
number_of_data = [i for i in range(1,unclean_data.shape[0]+1)]
plt.figure(figsize=(7,5))
plt.scatter(x= [unclean_data['latitude']/unclean_data['longitude']],y= unclean_data['price'])
plt.show()
```

```python
print(unclean_data.shape[0])
print(len([i for i in range(1,unclean_data.shape[0]+1)]))
```

```python
X = [[unclean_data['latitude'],unclean_data['longitude']]]
```

```python
X
```

```python
X = np.array(X)
X.shape
```

```python
X
```

```python
kmeans = KMeans(3)
```

```python
X_train = X.reshape(-1,2)
```

```python
X_train
```

```python
kmeans.fit(X_train)
```

```python
pred_ = kmeans.predict(X_train)
```

```python
pred_ 
```

```python
clustered_df = pd.DataFrame(X_train,columns = [ 'lat','long'])
```

```python
clustered_df['cluster'] = pred_
clustered_df.head()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(x= clustered_df['lat'],y=clustered_df['long'],c=clustered_df['cluster'])
plt.show()
```

```python
wcss =[]
for i in range(1,10):
    km = KMeans(i)
    km.fit(X_train)
    wcss_iter_  = km.inertia_
    wcss.append(wcss_iter_)
```

```python
number_of_cluster = [i for i in range(1,10)]
plt.plot(number_of_cluster,wcss)
plt.grid()
plt.show()
```
